```python
import pandas as pd
import unicodedata
import pyarrow
import fastparquet
```

```python
df_silver = pd.read_parquet('../data/2-silver/silver_acidentes.parquet')
df_gold = df_silver.copy()
```

```python
df_gold.dtypes
```

```python
colunas_utilizadas = [
    'data_inversa',
    'dia_semana',
    'horario',
    'uf',
    'municipio',
    'causa_acidente',
    'tipo_acidente',
    'classificacao_acidente',
    'fase_dia',
    'condicao_metereologica',
    'tipo_veiculo',
    'estado_fisico']
```

```python
df_gold = df_gold[colunas_utilizadas]
df_gold.dtypes
```

### Criando colunas

```python
#Criando colunas de tempo para facilitar a identificação no Power Bi
df_gold['ano'] = df_gold['data_inversa'].dt.year
df_gold['mes'] = df_gold['data_inversa'].dt.month
```

```python
mapa_dias = {
    'Segunda-feira': 1,
    'Terça-feira': 2,
    'Quarta-feira': 3,
    'Quinta-feira': 4,
    'Sexta-feira': 5,
    'Sábado': 6,
    'Domingo': 7
}
df_gold['dia_semana_num'] = df_gold['dia_semana'].map(mapa_dias)
df_gold.head(5)
```

```python
df_gold['tipo_veiculo'].unique()
```

```python
df_gold['tipo_veiculo'] = df_gold['tipo_veiculo'].str.lower().str.strip()

df_gold['tipo_veiculo'] = df_gold['tipo_veiculo'].replace({
    'não identificado': 'outros',
    'desconhecido': 'outros',
    '(null)': 'outros',
    'outro': 'outros',
    'outros': 'outros'
})
```

```python
df_gold['tipo_veiculo'] = df_gold['tipo_veiculo'].str.strip().str.title()
```

```python
mapa_categoria = {
    # Leves
    'Automóvel': 'Leve',
    'Caminhonete': 'Leve',
    'Camioneta': 'Leve',
    'Utilitário': 'Leve',
    'Utilitario': 'Leve',
    'Motor-casa': 'Leve',

    # Pesados
    'Caminhão': 'Pesado',
    'Caminhao trator': 'Pesado',
    'Caminhão-tanque': 'Pesado',
    'Ônibus': 'Pesado',
    'Microônibus': 'Pesado',
    'Reboque': 'Pesado',
    'Semi-reboque': 'Pesado',

    # 🏍️ Motos
    'Motocicletas': 'Moto',
    'Motoneta': 'Moto',
    'Miclomotor': 'Moto',
    'triciclo': 'Moto',
    'quadriciclo': 'Moto',

    # 🚜 Agrícola / especial (pode considerar pesado também)
    'Trator de rodas': 'Pesado',
    'Trator misto': 'Pesado',

    # 🚲 Não motorizados
    'Bicicleta': 'Não motorizado',
    'Carroça': 'Não motorizado',
    'Charrete': 'Não motorizado',
    'Carroça-charrete': 'Não motorizado',
    'Carro de mão': 'Não motorizado',

    # 🚆 Ferroviário
    'Bonde / trem': 'Ferroviário',
    'Trem-bonde': 'Ferroviário',

    # ❓ Outros / desconhecidos
    'outros': 'Outros'
}
```

```python
df_gold['categoria_veiculo'] = df_gold['tipo_veiculo'].map(mapa_categoria)
df_gold['tipo_veiculo'] = df_gold['tipo_veiculo'].str.title()
df_gold['categoria_veiculo'] = df_gold['categoria_veiculo'].fillna('Outros')
```
